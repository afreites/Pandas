## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins= pd.DataFrame(purchase_data)
age_bins.head()
```

```python
age_bins.groupby(["SN", "Age"]).size().reset_index(name='count')
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0,9.9,14.9, 19.9, 24.9,29.9,34.9,39.9,45]
```

```python

# Create labels for these bins
group_labels = ["< 10", "10 to 14","15 to 19", "20 to 24",
                "25 to 29", "30 to 34","35 to 39", "40+"]
```

```python
age_group=pd.DataFrame(age_bins.groupby(["SN", "Age"]).size().reset_index(name='count'))
age_group.head()
```

```python
pd.cut(age_group["Age"], bins, labels=group_labels).head()
```

```python
age_group["Age Bins"] = pd.cut(age_bins["Age"], bins, labels=group_labels)
age_group.head(10)
```

```python
# Create a GroupBy object based upon "View Group"
age_group_df = age_group.groupby("Age Bins")

type(age_group_df)
```

```python
age_summary= age_group_df[["count"]].count()
age_summary
```

```python
age_percent= age_summary["count"]/age_summary["count"].sum()
age_summary['age_percent']= age_percent
age_summary.style.format({"age_percent":'{:,.2%}'.format})
age_summary.rename(columns={'count': 'Counts','age_percent': 'Percentage'}, inplace=True)
```

```python
age_summary = age_summary.to_html(formatters={
    'Counts':'{:,.0f}'.format,
    'Percentage':'{:,.2%}'.format
})
display(HTML(age_summary))
```
